```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./FIFA22_preprocessed.csv", header=0, index_col=0)
df.head(5)
```

```python
df.info()
```

```python
df.describe()
```

```python
numeric_df = df.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numeric_df.corr()

variables_subsets = [
    ['Age', 'Value', 'International Reputation', 'Weak Foot', 'Skill Moves', 'Height'],
    ['Weight', 'Crossing', 'Finishing', 'HeadingAccuracy', 'ShortPassing', 'Volleys'],
    ['Dribbling', 'Curve', 'FKAccuracy', 'LongPassing', 'BallControl', 'Acceleration'],
    ['SprintSpeed', 'Agility', 'Reactions', 'Balance', 'ShotPower', 'Jumping'],
    ['Stamina', 'Strength', 'LongShots', 'Aggression', 'Interceptions', 'Positioning', ],
    ['Vision', 'Penalties', 'Composure', 'StandingTackle', 'SlidingTackle', 'DefensiveAwareness']
]

for i, subset_i in enumerate(variables_subsets):
    for j, subset_j in enumerate(variables_subsets):
        if j < i: continue
        plt.figure(figsize=(3.75,2))
        sns.heatmap(
            correlation_matrix.loc[subset_i, subset_j], 
            annot=True, cmap='coolwarm', fmt=".2f", 
            vmin=-1, vmax=1
            )
        plt.show()
```

```python
fig, axes = plt.subplots(6, 6, figsize=(20, 20))
axes = axes.flatten()

for i, subset in enumerate(variables_subsets):
    for j, elem in enumerate(subset):
        elem_df = df[[elem]]
        
        sns.histplot(elem_df, ax=axes[i*6 + j])

plt.tight_layout()
plt.show()
```

```python
dft = df.groupby("Best Position")["Preferred Foot"].value_counts().reset_index(name='Count')

p = sns.barplot(dft, x='Best Position', y='Count', hue='Preferred Foot')
```
