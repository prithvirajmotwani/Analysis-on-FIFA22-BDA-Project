```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Dataset taken from https://www.kaggle.com/datasets/bryanb/df-player-stats-database/data?select=df22_official_data.csv
df_full = pd.read_csv("./FIFA22_official_data.csv")
df_full.head(5)
```

```python
df_full.shape
```

```python
columns_to_drop = [
    'Photo', 'Flag', 'Overall', 'Potential', 'Club Logo', 'Wage', 'Special', 
    'Real Face', 'Position', 'Jersey Number', 'Joined', 'Loaned From', 
    'Contract Valid Until', 'Best Overall Rating', 'Release Clause'
]
df = df_full.drop(columns=columns_to_drop)

df.head()
```

```python
df.sample(20)[["Value", "Height", "Weight"]]
```

```python
(df.loc[df['Value'].str.startswith("€")].shape[0] ,
 
df.loc[df['Value'].str.endswith("M")].shape[0] + \
df.loc[df['Value'].str.endswith("K")].shape[0] ,

df.loc[df['Height'].str.endswith("cm")].shape[0] ,

df.loc[df['Weight'].str.endswith("kg")].shape[0])
```

```python
df['Value'] = df['Value'].str.split('€')
df['Value'] = df['Value'].apply(lambda x:x[1])

df_value_K = df.loc[df['Value'].str.endswith('K')]
df_value_K['Value'] = df_value_K['Value'].apply(lambda x: x[:-1])
df_value_K['Value'] = df_value_K['Value'].astype('float64')
df_value_K['Value'] = df_value_K['Value'] / 1000

df_value_M = df.loc[df['Value'].str.endswith('M')]
df_value_M['Value'] = df_value_M['Value'].apply(lambda x: x[:-1])
df_value_M['Value'] = df_value_M['Value'].astype('float64')

df = pd.concat([df_value_M,df_value_K])

df = df[df['Value'] != 0]

df['Height'] = df['Height'].apply(lambda x : x[:-2])
df['Height'] = df['Height'].astype('int64')

df['Weight'] = df['Weight'].apply(lambda x : x[:-2])
df['Weight'] = df['Weight'].astype('int64')

df.head()
```

```python
df.sample(30)[["Value", "Height", "Weight"]]
```

```python
df.shape
```

```python
df.info()
```

```python
gk_columns_to_drop = [
    'GKDiving', 'GKHandling', 'GKKicking', 
    'GKPositioning', 'GKReflexes'
]
df = df.drop(columns=gk_columns_to_drop)
df = df[df['Best Position'] != 'GK']

df.head()
```

```python
df.shape
```

```python
df = df.drop(columns=['Marking'])

df['Composure'].fillna(df['Composure'].mean(), inplace=True)
df['DefensiveAwareness'].fillna(df['DefensiveAwareness'].mean(), inplace=True)

df.isnull().sum()
```

```python
df.sample(5)
```

```python
df.shape
```

```python
df.info()
```

```python
df['Best Position'].unique()
```

```python
df.loc[df['Best Position'].isin(('CF', 'ST')),'Best Position'] = 'ST'
df.loc[df['Best Position'].isin(('LW', 'LM')),'Best Position'] = 'LM'
df.loc[df['Best Position'].isin(('RW', 'RM')),'Best Position'] = 'RM'
df.loc[df['Best Position'].isin(('LB', 'LWB')),'Best Position'] = 'LB'
df.loc[df['Best Position'].isin(('RB', 'RWB')),'Best Position'] = 'RB'
```

```python
df['Best Position'].unique()
```

```python
df.drop_duplicates(inplace=True)
df.shape
```

```python
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
df.to_csv('FIFA22_preprocessed.csv')
df.sample(20)
```
